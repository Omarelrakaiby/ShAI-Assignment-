# Task 1


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline 
%reload_ext autoreload
%autoreload 2


df = pd.read_csv('D:\Omar\Data\Salaries.csv')
# df = pd.read_csv(csv_path)

no_rows, no_columns = df.shape

d_types = df.dtypes

missing_values = df.isna().sum()
#missing means that the particular column is empty  
# Display the results
print("\n"+'Number of rows: '+str (no_rows))
print('Number of columns: '+str (no_columns))
print("\n"+"Data types:-")
print(d_types)
print("\n"+"Missing values:")
print(missing_values)

```

# Task 2

```python
Mean = df['TotalPay'].mean()
print("Mean is :",Mean)
Med = df['TotalPay'].median()
print("Median is :",Med)
Mode = df['TotalPay'].mode().iloc[0]
print("Mode is :",Mode)
Min = df['TotalPay'].min()
print("Minimum is :",Min)
Max = df['TotalPay'].max()
print("Maxiumum is :",Max)
Range = Max - Min
print("Range is :",Range)
Std = df['TotalPay'].std()
print("Standard deviation is :",Std)
```

# Task 3

```python
df = df.drop(columns = "Id")#to remove any column
df = df.drop(columns = "Notes")#to remove any column 
df = df.drop(columns = "Status")#to remove any column
df = df.drop(columns = "Benefits")#to remove any column
df = df.drop_duplicates()# to drop duplicates 
df = df.fillna('')
df
```

```python
def capitalize_name(name):
    return ' '.join(word.capitalize() for word in name.split())

# Find the index of the "employeeName" column
employee_name_index = df.columns.get_loc("EmployeeName")

# Capitalize the names in the "employeeName" column
df["EmployeeName"] = df["EmployeeName"].apply(capitalize_name)
for x in df.index:
    if df.loc[x,"JobTitle"] == 'Not provided':
        df.drop(x,inplace=True)#remove the ones that have yes in do not contact 

def capitalize_name(name):
    return ' '.join(word.capitalize() for word in name.split())

# Find the index of the "employeeName" column
employee_name_index = df.columns.get_loc("JobTitle")

# Capitalize the names in the "employeeName" column
df["JobTitle"] = df["JobTitle"].apply(capitalize_name)
for x in df.index:
    if df.loc[x,"TotalPay"] == 0.0:
        df.drop(x,inplace=True)#remove the ones that have yes in do not contact
df["BasePay"] = df["BasePay"].replace('','0',regex=True)
df["OvertimePay"] = df["OvertimePay"].replace('','0',regex=True)
df["OtherPay"] = df["OtherPay"].replace('','0',regex=True)
df["TotalPay"] = df["TotalPay"].replace('','0',regex=True)
df["TotalPayBenefits"] = df["TotalPayBenefits"].replace('','0',regex=True)
df
```

# Task 4

```python
bins = [0,100000,200000,300000,400000,500000,600000]
salary = df['TotalPay']

plt.hist(salary ,bins = 25 , color='#abcdef' ,edgecolor = 'black')

plt.xticks(bins)

plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries', fontdict={'fontweight':'bold','fontsize': 14})
#plt.savefig('DistrofSalaries.png',dpi=300)


plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['JobTitle'].value_counts().nlargest(14).plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
#plt.savefig('PieChartofsalaries.png',dpi=300)
plt.show()
```

# Task 5

```python
# Create groups of 50 rows each
group_size = 50
df['group'] = (df.index // group_size) + 1

# Calculate summary statistics for each group
grouped_stats = df.groupby('group')['TotalPay'].agg(['count', 'mean', 'median', 'std'])

# Print the summary statistics
print(grouped_stats)

# Compare average salaries across different groups
average_salaries = grouped_stats['mean']
print("Comparison of Average Salaries Across Groups:")
print(average_salaries)
```
